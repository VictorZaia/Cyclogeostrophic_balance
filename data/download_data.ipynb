```python
import xarray as xr
```

```python
dir_data = '/mnt/meom/workdir/albert/eNATL60/eNATL60-BLB002-S/1h/GULFSTREAM/'

date = 'y2009m08d01'


mode = 'lr' # either 'lr' for low resolution, or 'hr' of high resolution
```

```python
if mode=='lr':
    down = 10
if mode=='hr':
    down = 2
    
nameu = 'eNATL60GULFSTREAM-BLB002_'+date+'.1h_sozocrtx.nc'
namev = 'eNATL60GULFSTREAM-BLB002_'+date+'.1h_somecrty.nc'
nameh = 'eNATL60GULFSTREAM-BLB002_'+date+'.1h_sossheig.nc'
```

```python
u = xr.open_dataset(dir_data+nameu)
v = xr.open_dataset(dir_data+namev)
h = xr.open_dataset(dir_data+nameh)

ds = xr.merge((u,v,h),compat='override').isel({'time_counter':slice(0,1),'y':slice(0,u.y.size-u.y.size%down),
                                               'x':slice(0,u.x.size-u.x.size%down)})
ds = ds.coarsen(x=down).mean().coarsen(y=down).mean()
ds
```

```python
ds.to_netcdf('eNATL60GULFSTREAM-BLB002_' + date + '_' + mode + '.nc')
```
